```
import matplotlib.pyplot as plt
import numpy as np
```

```
%matplotlib inline
```

```
plt.style.use("seaborn-whitegrid")
```

```
fig = plt.figure()
ax   = plt.axes()

x = np.linspace(-10,10,10000)
ax.plot(x, np.sin(x))
```

```
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```
plt.plot(x, np.log(x))
plt.plot(x, np.log(-x))
```

```
plt.plot(x, np.exp(x))
plt.plot(x, np.exp(-x))
```

```
x  = np.linspace(-10,10,10000)

plt.plot(x, x**2, '-g', label = "bowl")
plt.plot(x, -(x**2), '--b', label = "inverted bowl")
plt.plot(x**2, x, '-.r', label = "standard parabola")
plt.plot(-(x**2), x, ':c', label = "inverted parabola")
plt.title("parabolic curve")
plt.xlabel("X")
plt.ylabel("Square of X")
plt.axis([-10,10,-10,10])
plt.legend()

plt.show()
```

```
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)

plt.scatter(x, y, c = colors, s = sizes, cmap = "viridis", alpha = 0.3)
plt.colorbar()
```
